```
import pandas as pd #importing for using dataframes
import gzip #importing for unzipping the file for reading it
import nltk
from collections import Counter #importing to use it to count 
from nltk.tokenize import MWETokenizer #importing MWE nltk 
from nltk.stem import WordNetLemmatizer #importing word net lemmetizer nltk 
from nltk import pos_tag
nltk.download('punkt')
nltk.download('wordnet')
nltk.download('averaged_perceptron_tagger')
```

```
data = pd.read_json("http://snap.stanford.edu/data/amazon/productGraph/categoryFiles/reviews_Office_Products_5.json.gz", lines = True) #Reading the Office Products file 
```

```
data.head(10) #viewing the first 10 rows of the data
```

```
df = data.copy() #making a copy of the data and storing it in 'df' to leave the original data unaltered.
```

```
line = df['reviewText'][0] #storing the string as 'line'
line
```

```
token = nltk.word_tokenize(line) #splitting line into tokens
token
```

```
mwe_tokenizer = nltk.tokenize.MWETokenizer() #MWE_Tokenizer
```

```
mwe = mwe_tokenizer.tokenize(line.split())
mwe
```

```
wnl = WordNetLemmatizer() #Woed_Net_Lemmatizer
```

```
lem = [wnl.lemmatize(word) for word in token]
lem
```

```
pos = nltk.pos_tag(line.split()) #Part_Of_Speech_Tag
pos
```

```
dic = {'Tokens':token,
        'MWETokenizer':mwe_tokenizer, 
        'WordNetLemmatizer':wnl,
        'POS_Tag':pos}
```

```
dic
```

```
adj = [] #Making a list of of the words which are only adjectives 
for dic in pos:
  if dic[1] == 'JJ':
    adj.append(dic[0])
adj
```

```
def pipeline(line): #Defining a pipeline to implement it on the entire columm
  pdic = {}
  token = nltk.word_tokenize(line)
  mwe = mwe_tokenizer.tokenize(line.split())
  lem = [wnl.lemmatize(word) for word in token]
  pos = nltk.pos_tag(line.split())
  pdic = {'Tokens':token,'MWETokenizer':mwe,'WordNetLemmatizer':lem,'POS_Tag':pos}
  return pdic
```

```
pipeline(line)
```

```
data['Pipeline_ReviewText'] = data.reviewText.apply(pipeline)
```

```
data['Pipeline_ReviewText']
```

```
data['Summmary'] = data.summary.apply(pipeline)
```
